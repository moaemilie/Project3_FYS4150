## RK4 in z direction

Single particle

```python
import matplotlib.pyplot as plt

x = []
y = []
z = []
for line in open('Particle_z_RK4.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(float(lines[0]))
    y.append(float(lines[1]))
    z.append(float(lines[2]))

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.set_title('One Particles movement in 50Î¼s', fontsize=15);
ax.set_xlabel('Posision [x]', fontsize=15)
ax.set_ylabel('Posision [y]', fontsize=15)
ax.set_zlabel('Posision [z]', fontsize=15)
ax.plot3D(x, y, z, 'gray')
plt.savefig('figure5.pdf')
```

```python
import numpy as np
time = np.linspace(0.0, 50.0, num=501)

fig = plt.figure(figsize=(8, 6))
plt.title('One Particle', fontsize=15)
plt.xlabel('time [Î¼s]', fontsize=15)
plt.ylabel('posision [z]', fontsize=15)
plt.plot(time[0:500], z)
plt.savefig('figure.pdf')
plt.show()
```
