## Euler in z direction

```python
import matplotlib.pyplot as plt
  
x = []
y = []
for line in open('Particle_z.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(float(lines[0]))
    y.append(float(lines[1]))
      
fig = plt.figure()
plt.title('Particle 1', fontsize=15)
plt.xlabel('time', fontsize=15)
plt.ylabel('posision_z', fontsize=15)
plt.plot(x)
  
plt.show()
```

```python
fig = plt.figure()
plt.title('Particle 2', fontsize=15)
plt.xlabel('time', fontsize=15)
plt.ylabel('posision_z', fontsize=15)
plt.plot(y)
  
plt.show()
```
