## RK4 in z direction

```python
import matplotlib.pyplot as plt

x = []
y = []
z = []
for line in open('Particle_z_RK4.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(float(lines[0]))
    y.append(float(lines[1]))
    z.append(float(lines[2]))
```

```python
fig = plt.figure(figsize=(12, 10))
#plt.title('F', fontsize=15)
plt.xlabel('Period[w_v]', fontsize=15)
plt.ylabel('Fraction of Particles left', fontsize=15)
plt.plot(p0_fine_1, f0_fine_1, label = "f = 0.1")
plt.plot(p0_fine_4, f0_fine_4, label = "f = 0.4")
plt.plot(p0_fine_7, f0_fine_7, label = "f = 0.7")
plt.title('Fraction of particles left in the Penning Trap after 500 micro seconds', fontsize=15)
plt.legend()
plt.savefig('Fraction_of_part_left_fine.pdf')
plt.show()
```
